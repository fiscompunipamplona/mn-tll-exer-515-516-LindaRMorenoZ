El error del método de diferencias hacía adelante es

# Ea = (2*c*|f(x)|/h) + (h*|f''(x)|/2) 

y para el método de diferencias centrales es

# Ec = (2*c*|f(x)|/h) + ((h^2)*|f'''(x)|/24)

como vemos el primer término de las dos formulas es igual, así que nos vamos a concentrar en el segundo término para ver la dependencia de h, |f''(x)| y |f'''(x)| depende de cada caso en partícular; para encontrar el valor general de h que cumpla la condición Ea < Ec vamos a usar los siguientes factores 

Para Ea 
# Ea=h/2

y para Ec
#Ec=(h^2)/24

```
Ea=2
Ec=1
h2=0
c=1e-16
while Ea >= Ec:
  h2 += 0.5
  Ea=(h2/2)
  Ec=(h2**2)/24
  print(Ea, Ec, h2)  

```
