```
import numpy
from math import tanh, cosh
from pylab import plot, linspace, legend

print('librerias')
```

```
def f(x):                                #def. función dependiente x
  return 1+((1/2)*tanh(2*x))

def df(x):                               #def. función derivada con respecto a x
  h = 1e-5                               # tamaño de paso
  return (f(x + 0.5 * h) - f(x - 0.5 * h)) / h

def g(x):                                 #def. derivada de f analíticamente
  return 1 /(cosh(2*x))** 2

x = linspace(-2, 2, 100)                  #intervalo de x, 100 pasos de -2 a 2 
dF = list(map(df, x))                     #valores para df evaluados en x
G = list(map(g, x))                       #valores para g evaluados en x

plot(x, dF, 'o', label= 'númerica')       #grafica los resultados, df y g vs x
plot(x, G, label='analítica')

legend()
```
